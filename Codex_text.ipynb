```python
from herbie import Herbie
from herbie.toolbox import EasyMap, pc
from herbie import paint
import matplotlib.pyplot as plt
import xarray as xr
```

```python
hrrr = Herbie(
    "2025-09-25 12:00", 
    model="hrrr", 
    product="sfc", 
    fxx=6
)

rtma = Herbie("2025-09-25 18:00", 
           model="rtma", 
           product="anl"
)
```

```python
ds_hrrr = hrrr.xarray("TMP:2 m above")
ds_hrrr.coords

ds_rtma = rtma.xarray("TMP:2 m above")
ds_rtma.coords
```

```python
# 3. Interpolate RTMA to the HRRR grid
rtma_on_hrrr = ds_rtma.interp(x=ds_hrrr.x, y=ds_hrrr.y)

# 4. Copy HRRR lat/lon variables so the metadata matches the new grid
rtma_on_hrrr["latitude"] = ds_hrrr["latitude"]
rtma_on_hrrr["longitude"] = ds_hrrr["longitude"]
```
