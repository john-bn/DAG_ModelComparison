```python
from herbie import Herbie
from herbie.toolbox import EasyMap, pc
from herbie import paint
import matplotlib.pyplot as plt
import xarray as xr
```

```python
hrrr = Herbie(
    "2025-09-25 12:00", 
    model="hrrr", 
    product="sfc", 
    fxx=6
)

rtma = Herbie("2025-09-25 18:00", 
           model="rtma", 
           product="anl"
)
```

```python
def plot_temp_diff(hrrr, rtma):
    # Extract 2-meter temperature data
    hrrr_temp = hrrr.get_variable("TMP:2 m")
    rtma_temp = rtma.get_variable("TMP:2 m")

    # Calculate the difference
    temp_diff = hrrr_temp - rtma_temp

    # Plotting
    plt.figure(figsize=(10, 6))
    ax = plt.axes(projection=pc())
    temp_diff.plot.pcolormesh(
        ax=ax,
        cmap="coolwarm",
        cbar_kwargs={"label": "Temperature Difference (K)"}
    )
    paint.add_coastlines(ax)
    paint.add_states(ax)
    paint.add_counties(ax)
    plt.title("2-Meter Temperature Difference: HRRR - RTMA\n2025-09-25 12:00 UTC (HRRR) vs 2025-09-25 18:00 UTC (RTMA)")
    plt.show()
```
